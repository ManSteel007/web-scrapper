To Extract it All - Single Page

```python
company = soup.find_all('div', class_= 'companyCardWrapper')
```

```python
len(company)
```

```python
name, ratings, reviews, ctype, hq, merits, demerits = [], [], [], [], [], [], []

for i in company:

    # for finding the name of the company
    name.append(i.find('h2', class_= 'companyCardWrapper__companyName').text.strip())

    # for finding the ratings of the company
    ratings.append(i.find('div', class_= 'rating_text rating_text--md').text.strip())

    # for finding the number of reviews of the company
    reviews.append(i.find('span', class_= 'companyCardWrapper__companyRatingCount').text.strip())

    # for finding the industry and headquarters of the company
    temp1=[]
    temp1.append(i.find('span', class_= 'companyCardWrapper__interLinking').text.strip())
    a,b = temp1[0].split('|', 1)
    ctype.append(a.strip())
    hq.append(b.strip())
    # temp2 = []
    # temp2.append(i.find('span', class_= 'companyCardWrapper__ratingValues').text.strip())

    # for finding the highly rated and most critically rated for of the company
    high_rate = []
    low_rate = []
    high = i.find('span', class_= 'companyCardWrapper__ratingHeader--high')
    if high:
        h_text = high.find_parent('div').find_next_sibling('span', class_= 'companyCardWrapper__ratingValues').text.strip()
        high_rate.append(h_text)
    else:
        high_rate.append(np.nan)
    merits.append(high_rate)
    low = i.find('span', class_= 'companyCardWrapper__ratingHeader--critical')
    if low:
        l_text = low.find_parent('div').find_next_sibling('span', class_= 'companyCardWrapper__ratingValues').text.strip()
        low_rate.append(l_text)
    else:
        low_rate.append(np.nan)
    demerits.append(low_rate)
```

Creating the data frame

```python
d = {'name': name, 'rating': ratings, 'reviews': reviews, 'company_type' : ctype, 'headquarters' : hq, 'most_rated': merits, 'most_critised': demerits}

df = pd.DataFrame(d)

df
```

Creating the dataset by extracting the above company information from 500 pages

```python
final = pd.DataFrame()
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

for j in range(1, 501):
    webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j), headers=headers).text
    soup=BeautifulSoup(webpage,'html.parser')
    company = soup.find_all('div', class_= 'companyCardWrapper')

    name, ratings, reviews, ctype, hq, merits, demerits = [], [], [], [], [], [], []

    for i in company:

        # for finding the name of the company
        try:
            name.append(i.find('h2', class_= 'companyCardWrapper__companyName').text.strip())
        except:
            name.append(np.nan)

        # for finding the ratings of the company
        try:
            ratings.append(i.find('div', class_= 'rating_text rating_text--md').text.strip())
        except:
            ratings.append(np.nan)

        # for finding the number of reviews of the company
        try:
            reviews.append(i.find('span', class_= 'companyCardWrapper__companyRatingCount').text.strip())
        except:
            reviews.append(np.nan)

        # for finding the industry and headquarters of the company
        try:
            temp1=[]
            temp1.append(i.find('span', class_= 'companyCardWrapper__interLinking').text.strip())
            a,b = temp1[0].split('|', 1)
            ctype.append(a.strip())
            hq.append(b.strip())
        except:
            ctype.append(np.nan)
            hq.append(np.nan)

        # for finding the highly rated and most critically rated for of the company
        try:
            high_rate = []
            low_rate = []
            high = i.find('span', class_= 'companyCardWrapper__ratingHeader--high')
            if high:
                h_text = high.find_parent('div').find_next_sibling('span', class_= 'companyCardWrapper__ratingValues').text.strip()
                high_rate.append(h_text)
            else:
                high_rate.append(np.nan)
            merits.append(high_rate)
            low = i.find('span', class_= 'companyCardWrapper__ratingHeader--critical')
            if low:
                l_text = low.find_parent('div').find_next_sibling('span', class_= 'companyCardWrapper__ratingValues').text.strip()
                low_rate.append(l_text)
            else:
                low_rate.append(np.nan)
            demerits.append(low_rate)
        except:
            continue

    df=pd.DataFrame({
        'Name' : name,
        'Rating' : ratings,
        'Reviews' : reviews,
        'Company_Type' : ctype,
        'Headquarters' : hq,
        'Highly Rated' : merits,
        'Most Critically Rated' : demerits
    })

    final = pd.concat([final, df], ignore_index=True)

    #Logging
    print('Data from page {} has been added in the dataset successfully'. format(j))
```
